```python
# ï‚· Add a new column named GRE_to_TOEFL_Ratio = GRE Score / TOEFL Score.
df['GRE_to_TOEFL_Ration'] = df['GRE Score'] / df['TOEFL Score']
df.head()
```

```python
# ï‚· Create a new column Admission_Chance_% = Chance of Admit Ã— 100.
df['Admission_Chance_%'] = df['Chance of Admit '] * 100
df['Admission_Chance_%']
```

```python
# ï‚· Rename the column Chance of Admit to Admit_Probability.
df.rename(columns = {'Admit_Probability': 'Chance of Admit'} , inplace= True)
df.columns
```

```python
# Drop the column Serial No..
df.drop(columns = ['Serial No.'] , inplace=True)
df.columns
```

```python
# Round all values in CGPA to one decimal place.
def rounding(x):
    return x[:2]
df['CGPA'] = df['CGPA'].round(1)
df['CGPA']
# df['CGPA'].dtype
```

```python
# df = pd.read_csv('Admission_Predict.csv')
df.head()
```

```python
# ï‚· Replace all University Rating values of 5 with 10.
def f(x):
    if x == 5:
         return 10
    else:
        return x        
df['University Rating'] = df['University Rating'].apply(f)
```

```python
df['University Rating']
```

```python
df['Research'].unique
```

```python
# Convert Research column values from 0/1 to â€œNoâ€/â€œYesâ€.
def convert_val(x):
    if x == 0:
        return 'No'
    else:
        return 'Yes'
df['Research'] = df['Research'].apply(convert_val)
```

```python
df['Research']
```

```python
# ï‚· Create a new column GRE_Level = â€œHighâ€ if GRE â‰¥ 320 else â€œLowâ€.
def new_gre(x):
    if x >= 320:
        return 'High'
    else:
        return 'Low'
df['GRE_Level'] = df['GRE Score'].apply(new_gre)
```

```python
df.head()
```

```python
# ï‚· Normalize CGPA column to range 0â€“1 using min-max scaling (without sklearn).
def normalize(x):
    min_gpa = df['CGPA'].min()
    max_gpa = df['CGPA'].max()
    if x == min_gpa:
        return 0
    elif x == max_gpa:
        return 1
    else:
        return (x- min_gpa) / (max_gpa - min_gpa)
df['CGPA'] = df['CGPA'].apply(normalize)
df['CGPA']
```

```python
# Increase all TOEFL Score values by 2 points.
def increse_toefl(x):
    return x + 2
df['TOEFL Score'] = df['TOEFL Score'].apply(increse_toefl)
```

```python
df['TOEFL Score']
```

```python
# ï‚· Add a column Weighted_Score = GRE Score * 0.3 + TOEFL Score * 0.2 + CGPA * 0.5.
df['Weighted_Score'] = (df['GRE Score'] * 0.3) + ( df['TOEFL Score'] * 0.2) + (df['CGPA'] * 0.5)
df['Weighted_Score']
```

```python
df.shape[0]
```

```python
# ï‚· Drop all duplicate rows if any.
df.drop_duplicates()
```

```python
# Replace all SOP values less than 2 with the mean SOP.
def replace_sop(x):
    sop_mean = df['SOP'].mean()
    if x == 2:
        return sop_mean
    else:
        return x
df['SOP'] = df['SOP'].apply(replace_sop)
df['SOP']
```

```python
df.info()
```

```python
# ï‚· Create a new column combining University Rating and Research (e.g.,&quot;Rating-Research&quot;).
df['University Rating'] = df['University Rating'].astype(str)
df['Combine Rating_Research'] = df['University Rating'] + df['Research']
df['Combine Rating_Research']
```

```python
df.columns
```

```python
# ï‚· Convert Chance of Admit to categorical bins: â€œLowâ€, â€œMediumâ€, â€œHighâ€.
def convert_chance_of_admit(x):
    avg = df['Chance of Admit '].mean()
    low = df['Chance of Admit '].min()
    high = df['Chance of Admit '].max()
    if x <= low:
        return 'Low'
    elif x >= low and x <= avg:
        return 'Medium'
    elif x >= avg and x <= high:
        return "High"
        
df['Chance of Admit '] = df['Chance of Admit '].apply(convert_chance_of_admit)
```

```python
df['Chance of Admit ']
```

```python
# ï‚· Add a constant column Country with value â€œIndiaâ€.
df['Country'] = 'India'
df['Country']
```

```python
df.columns
```

```python
# ï‚· Reorder columns so that Chance of Admit comes first.
new_col = ['Chance of Admit ','GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA',
           'Research','GRE_to_TOEFL_Ration','Admission_Chance_%', 'GRE_Level', 'Weighted_Score',
           'Combine Rating_Research', 'Country']
df = df.reindex(columns = new_col)
```

```python
df.head()
```

```python
# ï‚· Replace values in LOR column where rating is 5 with 4.9.
def replace_lop(x):
    if x == 5:
        return 4.9
    else:
        return x
df['LOR '] = df['LOR '].apply(replace_lop)
df['LOR ']
```

```python
# ï‚· Convert all float columns to integers (after rounding).
for col in df.columns:
    if df[col].dtype == 'float':
        df[col] = df[col].round().astype('int')
```

```python
df.info()
```

```python
# ï‚· Replace CGPA values above 9.5 with 9.5.
def replace_cgpa(x):
    if x >= 9.5:
        return 9.5
    else:
        return x
df['CGPA'] = df['CGPA'].apply(replace_cgpa)
df['CGPA']
```

```python
df.columns
```

```python
# Apply a lambda function to square all SOP values.
df['SOP'] = df['SOP'].round().apply(lambda x:x*x)
```

```python
# ï‚· Add a column GRE_Category = â€œBelow Avgâ€, â€œAvgâ€, â€œAbove Avgâ€ based on mean GRE.
def avg_gre(x):
    avg = df['GRE Score'].mean()
    if x < avg:
        return "Below Avg"
    elif x == avg:
        return 'Avg'
    elif x > avg:
        return "Above Avg"
df['GRE_Category'] = df['GRE Score'].apply(avg_gre)
df['GRE_Category']
```

```python
# ï‚· Standardize column names to snake_case (lowercase, replace spaces with _).

df.columns = [x.lower() for x in df.columns]
df.columns = [x.replace(' ', '_') for x in df.columns]
```
