# 2. Column Operations &amp; Transformations (25 Questions).`

```python
# 26. Add a new column overall_avg = (ssc_p + hsc_p + degree_p + mba_p) / 4.
df['overall_avg'] = df['ssc_p'] + df['hsc_p'] + df['degree_p'] + df['mba_p'] / 4
```

```python
df['overall_avg']
```

```python
# 27. Rename the column degree_p to degree_percentage.
df.rename(columns = {'degree_p' :'degree_percentage'})
```

```python
# 28. Create a new column placement_status = “Hired” if status = “Placed” else “Not Hired”.
def f(x):
    if x == 'Placed':
        return 'Hired'
    else:
        return 'Not hired'
df['placement_status'] = df['status'].apply(f)
```

```python
df['placement_status']
```

```python
# 29. Drop the column sl_no.
df.drop(columns = 'sl_no')
```

```python
# 30. Replace all “Comm&amp;Mgmt” in degree_t with “Commerce”.
df['degree_t'].replace('Comm&Mgmt' , 'Commerce' ,  inplace=True )
```

```python
df['degree_t']
```

```python
# 31. Convert gender column values “M” → “Male”, “F” → “Female”.
def convert(x):
    if x == 1:
        return "Male"
    else:
        return "Female"
df['gender'] = df['gender'].apply(convert)
```

```python
df
```

```python
# 32. Create a new column above_80_ssc = True if ssc_p &gt; 80 else False.
df['above_80_ssc'] = df['ssc_p'].apply(lambda x:True if x>80 else False)
```

```python
df['above_80_ssc']
```

```python
df.head()
```

```python
# 33. Round all percentage columns to 1 decimal place.
cols = ['ssc_p', 'hsc_p', 'degree_p' , 'etest_p', 'mba_p']
df[cols] = df[cols].round(1)
```

```python
df.head()
```

```python
# 34. Convert all column names to lowercase.
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
# 35. Add a new column salary_in_lpa = salary / 100000.
df['salary_in_lpa'] = df['salary'] / 100000
```

```python
df['salary_in_lpa']
```

```python
# 36. Replace NaN values in salary with 0.
df['salary'].fillna(0 , inplace=True)
```

```python
df['salary']
```

```python
# 37. Drop all duplicate rows from the dataset.
df.drop_duplicates()
```

```python
# 38. Increase all mba_p scores by 5 points.
df['mba_p'] = df['mba_p'] + 5
```

```python
df['mba_p']
```

```python
# 39. Convert salary column to integer type.
df['salary'] = df['salary'].astype(int)
```

```python
df.columns
```

```python
# 40. Reorder columns to make status the first column.
reorder = df.loc[:, ['sl_no', 'status', 'gender', 'ssc_p', 'ssc_b', 'hsc_p', 'hsc_b', 'hsc_s',
       'degree_p', 'degree_t', 'workex', 'etest_p', 'specialisation', 'mba_p',
       'salary', 'overall_avg', 'placement_status', 'above_80_ssc',
       'salary_in_lpa']]
reorder
```

```python
df
```

```python
# 41. Replace all “Yes”/“No” in workex with 1/0.
df['workex'] = df['workex'].apply(lambda x:1 if 'Yes' in x else 0)
```

```python
df['workex']
```

```python
# 42. Create a new column performance_tag: “Excellent” if degree_p &gt; 80 else “Average”.
def fx(x):
    if x > 80:
        return "Exellent"
    else:
        return 'Average'
df['performance_tag'] = df['degree_p'].apply(fx)
```

```python
# 43. Standardize all column names to snake_case (e.g., ssc_p → ssc_percentage).
df.columns = [
    col.replace('_p' , '_percentage') if col.endswith('_p') else col
    for  col in df.columns
]
```

```python
df
```

```python
# 44. Create a new column combining degree_t and specialisation (e.g., “Sci&amp;Tech_Mkt&amp;HR”).
df['combine_degreeandspel'] = df['degree_t'] + df['specialisation']
```

```python
df
```

```python
# 45. Fill missing etest_p values with mean of that column.
df['etest_percentage'].fillna(df['etest_percentage'].mean() , inplace=True)
```

```python
df['etest_percentage']
```

```python
# 46. Create a new column total_score = ssc_p + hsc_p + degree_p + mba_p.
df['total_score'] = df['ssc_percentage'] + df['hsc_percentage'] + df['degree_percentage'] + df['mba_percentage']
```

```python
df['total_score']
```

```python
# 49. Add a column mba_grade = “A” if mba_p ≥ 75 else “B”.
df['mba_grade'] = df['mba_percentage'].apply(lambda x:'A' if x>= 70 else 'B')
```
