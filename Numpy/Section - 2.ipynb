### SECTION 2 â€” Array Creation Techniques

```python
import numpy as np
```

```python
# 1. np.zeros()
# Creates an array filled with 0.
zeros = np.zeros((3,3))
zeros
```

```python
# 2. np.ones()
# Creates an array filled with 1.
ones = np.ones((2,4))
ones
```

```python
# 3. np.arange()
# Creates a range of values.
arange = np.arange(1, 10, 2)    # start = 1, stop = 10 , step = 2
arange
```

```python
# 4. np.linspace()
# Creates evenly spaced numbers between a range.
linspace = np.linspace(0,1,5)
linspace
```

```python
# 5. np.eye()
# Creates an identity matrix.
# identity matrix means all principle diagonal elements are 1s and other are zero
eye = np.eye(4)
eye
```

```python
# 6. Random Matrices
# Random integers:
random = np.random.randint(1,100, (3,3))   # from 1 to 100 and 3,3 matrix
random
```

```python
# Random normal distribution: 
random_dist = np.random.randn(3,3)
random_dist
```
